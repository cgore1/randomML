```python
batsman= gl.SFrame("batsmanClusters.csv")
historical=gl.SFrame("historical/HistoricalFeatures.csv")
```

```python
segments=[]
for i in range(10):
    seg=gl.SFrame("segments//segment"+str(i+1)+".csv")
    segments.append(seg)
```

```python
all_features=[]
for i in range(10):
    combined=historical.join(segments[i], on='Team')
    combined= combined.join(batsman, on='Batsman')
    all_features.append(combined)
```

```python
modelFeatures=[]
for feature in all_features:
    #mdFeature=feature.select_columns(['Runs Scored','Wickets Lost','Got All Out', 'Runs Conceded','Opponent Wickets Taken','Opponent All Out','Total Runs Scored','Balls Faced', 'R0','W0','R1','W1','Target','ClusterID','Non Home Runs','Home','Current_NHR'])
    mdFeature=feature.select_columns(['Runs Scored','Wickets Lost','Total Runs Scored','Balls Faced', 'R0','W0','R1','W1','Target','ClusterID','Non Home Runs','Home','Current_NHR'])
    modelFeatures.append(mdFeature)
```

```python
models=[]
for i in range(10):
    linearModel= gl.linear_regression.create(modelFeatures[i], target='Current_NHR', validation_set=None)
    models.append(linearModel)
```
